```python
import pandas as pd
import numpy as np
from datetime import datetime
import sqlite3
```

```python
conn = 'database.sqlite'
database=sqlite3.connect(conn)
table = pd.read_sql('''SELECT * FROM sqlite_master WHERE
        type='table';''',database)
table
```

```python
match_details = pd.read_sql('''SELECT Season_Id, Match_Id,

v.Venue_Name, c.City_Name, t.Team_Name AS Winner

FROM Match

INNER JOIN Venue AS v ON match.Venue_Id == v.Venue_Id

INNER JOIN City AS c ON v.City_Id == c.City_Id

INNER JOIN Team AS t ON match.Match_Winner == t.Team_Id;''', database)

match_details
```

```python
teams=pd.read_sql('''SELECT * FROM Team;''',database)
teams
```

```python
season=pd.read_sql('''SELECT * FROM Season;''',database)
season
```

```python
match_details_PW=pd.read_sql('''SELECT Match_Id,Team_2 as Away_Team,
            Toss_Winner, Match_Winner FROM Match WHERE Team_1=(SELECT
            Team_1 FROM Match WHERE Team_1 == 10 AND Season_Id==5)''',
            database)
match_details_PW
```
